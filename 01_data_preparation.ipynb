```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

plt.style.use(['seaborn', "seaborn-dark"][0])
plt.rc('axes', titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rcParams.update({'font.size': 16})
plt.rcParams['axes.titlesize'] = 16
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams.update({'lines.markeredgewidth': 1})
plt.rcParams.update({'errorbar.capsize': 2})

import pickle
from sklearn.datasets import make_moons, make_circles, make_classification
```

# Datasets

```python
def plot_dataset(X, y, name):
    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap="RdYlBu");
    ax.set_title(f"{name} dataset");
```

```python
dataset_spiral = make_moons(n_samples=100, noise=0.1, random_state=0)
plot_dataset(dataset_spiral[0], dataset_spiral[1], "Espiral")
# plt.savefig("dt1.pdf", bbox_inches='tight', )
```

```python
dataset_spiral_test = make_moons(n_samples=80, noise=0.1, random_state=1)
plot_dataset(dataset_spiral_test[0], dataset_spiral_test[1], "Espiral")
```

```python
dataset_donut = make_circles(n_samples=100, noise=0.1, factor=0.2, random_state=1)
plot_dataset(dataset_donut[0], dataset_donut[1], "Dona")
# plt.savefig("dt2.pdf", bbox_inches='tight', )
```

```python
dataset_donut_test = make_circles(n_samples=80,noise=0.1, factor=0.2, random_state=2)
plot_dataset(dataset_donut_test[0], dataset_donut_test[1], "Dona")
# plt.savefig("dt2.pdf", bbox_inches='tight', )
```

```python
X, y = make_classification(
    n_samples=100, n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1
)
X += 1.5 * np.random.RandomState(0).uniform(size=X.shape)
linearly_separable = (X, y)
plot_dataset(linearly_separable[0], linearly_separable[1], "Linear")
```

```python
X, y = make_classification(
    n_samples=100, n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1
)
X += 1.5 * np.random.RandomState(1).uniform(size=X.shape)
linearly_separable_test = (X[:80], y[:80])
plot_dataset(linearly_separable_test[0], linearly_separable_test[1], "Linear")
```

```python
X = 1-2*np.random.RandomState(0).uniform(size=(100, 2))
y = ((X[:, 0] * X[:, 1]) > 0).astype(int)
dataset_uniform = (X, y)
plot_dataset(dataset_uniform[0], dataset_uniform[1], "XOR")
```

```python
X = 1-2*np.random.RandomState(1).uniform(size=(80, 2))
y = ((X[:, 0] * X[:, 1]) > 0).astype(int)
dataset_uniform_test = (X, y)
plot_dataset(dataset_uniform_test[0], dataset_uniform_test[1], "XOR")
```
