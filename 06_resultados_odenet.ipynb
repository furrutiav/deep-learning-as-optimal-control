# Control

```python
def dynamic(control, X, alpha=1):
    O = []
    T = []
    for k in range(X.shape[0]):
        x0 = X[k]
        xn = x0.copy()
        x = [xn]
        for i in range(num_layers):
            ix = (dimension*(dimension+1)+1)*i
            jx = ix+dimension*dimension
            zx = jx+dimension
            K = control[ix:jx]
            be = control[jx:zx]
            Dt = control[zx]
            KK = np.array([
                [K[0], K[1]],
                [K[2], K[3]]
            ])
            xn = (xn + alpha*Dt*activation_function(KK@xn + be)).copy()
            xn = xn
            x.append(xn)

        ix = dimension*(dimension+1)
        jx = -ix+dimension*dimension
        W = control[-ix:jx]
        mu = control[jx:]
        WW = np.array([
            [W[0], W[1]],
            [W[2], W[3]]
        ]) 
        xf = (WW@xn + mu).copy()
        x.append(xf)
        T.append(x)
        sxf = np.exp(xf)
        sxf /= sum(sxf)
        o = sxf[1]
        O.append(o)
    O = np.array(O)
    return O, T
```

```python
def obj(control, X, y, alpha=1):
    O, _ = dynamic(control, X, alpha)
    L = y[:O.shape[0]]
    Dts = []
    for i in range(num_layers):
        ix = (dimension*(dimension+1)+1)*i
        jx = ix+dimension*dimension
        zx = jx+dimension
        Dt = control[zx]
        Dts.append(Dt)
    Dts = np.array(Dts)
    return 0.5*sum((O-L)*(O-L)) + 0.01*sum((1-2*((Dts>0)*(Dts<1)))>0)
```

```python
def get_cls(control):
        ix = dimension*(dimension+1)
        jx = -ix+dimension*dimension
        W = control[-ix:jx]
        mu = control[jx:]
        WW = np.array([
            [W[0], W[1]],
            [W[2], W[3]]
        ]) 
        return WW, mu
```

```python
def clf_cls(x, WW, mu):
    o = np.zeros(x.shape[0])
    for i in range(x.shape[0]):
        xf = WW@x[i]+mu
        sxf = np.exp(xf)
        sxf /= sum(sxf)
        o[i] = sxf[1]
    return o
```

```python
def clf(x, control, alpha):
    O, _ = dynamic(control, x, alpha)
    return O
```

```python
def plot_iterations(num_layers, T, y, WW, mu):
    fig, ax = plt.subplots(1, num_layers+1, figsize=(6*(num_layers+1), 5))
    h=0.1
    for i in range(0,num_layers+1):
        Xi = np.array(T)[:, i, :]
        x_min, x_max = Xi[:,0].min()-1, Xi[:,0].max()+1
        y_min, y_max = Xi[:,1].min()-1, Xi[:,1].max()+1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))
        Z = clf_cls(np.c_[xx.ravel(), yy.ravel()], WW, mu)
        Z = Z.reshape(xx.shape)
        ax[i].contourf(xx, yy, Z, cmap="RdYlBu", alpha=0.5)
        ax[i].scatter(Xi[:,0], Xi[:,1], c=y, cmap="RdYlBu")
        ax[i].set_xticks([])
        ax[i].set_yticks([]);
        ax[i].set_title(f"Iteration: {i}");
```

```python
def plot_region(X, y, control, alpha=1, model=""):
    fig, ax = plt.subplots(1, 1, figsize=(6*1, 5))
    h=0.1
    Xi = X
    x_min, x_max = Xi[:,0].min()-1, Xi[:,0].max()+1
    y_min, y_max = Xi[:,1].min()-1, Xi[:,1].max()+1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf(np.c_[xx.ravel(), yy.ravel()], control, alpha)
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap="RdYlBu", alpha=0.5)
    ax.scatter(Xi[:,0], Xi[:,1], c=y, cmap="RdYlBu")
    ax.set_xticks([])
    ax.set_yticks([]);
    ax.set_title(model, fontsize=24);
    ax.set_ylabel("Prediccin", fontsize=24)
```

```python
def plot_transformation(T, y, WW, mu, model=""):
    fig, ax = plt.subplots(1, 1, figsize=(6*1, 5))
    h=0.1
    i = -2
    Xi = np.array(T)[:, i, :]
    x_min, x_max = Xi[:,0].min()-1, Xi[:,0].max()+1
    y_min, y_max = Xi[:,1].min()-1, Xi[:,1].max()+1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))
    Z = clf_cls(np.c_[xx.ravel(), yy.ravel()], WW, mu)
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap="RdYlBu", alpha=0.5)
    ax.scatter(Xi[:,0], Xi[:,1], c=y, cmap="RdYlBu")
    ax.set_xticks([])
    ax.set_yticks([]);
    ax.set_title(model, fontsize=24);
    ax.set_ylabel("Transformacin", fontsize=24)
```

# Resultados

```python
dimension = 2
num_layers = 5
activation_function = lambda x: np.tanh(x)
dt = 1/num_layers
```

```python
control_0 = []
for i in range(num_layers):
    K = [1,0,0,1]
    be = [0,0]
    Dt = [dt]
    control_0 += K+be+Dt
W = [1,0,0,1]
mu = [0,0]
control_0 += W+mu

control_0 = np.array(control_0)
control_0;
```

```python
%%time
alpha=0.15
X, y = dataset_spiral_train
sol_spiral = minimize(obj, x0=control_0, args=(X, y, alpha))
optimal_control_spiral = sol_spiral.x
O_spiral, T_spiral = dynamic(optimal_control_spiral, X, alpha)
WW_spiral, mu_spiral = get_cls(optimal_control_spiral)
```

```python
plot_iterations(num_layers, T_spiral, dataset_spiral_train[1], WW_spiral, mu_spiral)
```

```python
plot_region(dataset_spiral_test[0], dataset_spiral_test[1], optimal_control_spiral, alpha=alpha, model="ODENet")
```

```python
plot_transformation(T_spiral, dataset_spiral_train[1], WW_spiral, mu_spiral, model="ODENet")
```
